# Stock Sentiment Analysis using News Headlines

```python
import pandas as pd 
```

```python
df= pd.read_csv('Data.csv')
```

```python
df.head()# data is from 3-Jan-2000 to 1-Jul-2016
```

```python
#preapring train and test
#data before 2015 is taken for training and data after 2015 is used for test.

train = df[df['Date']<'20150101']
test = df[df['Date']>'20141231']
```

```python
#Removing punctuations
data = train.iloc[:,2:27]
data.replace("[^a-zA-Z]"," " , regex=True, inplace=True)
```

```python
data.sample(3)
```

```python
#Renaming columns to make easy access
list = [i for i in range(25)]
new_index = [str(i) for i in list]
data.columns = new_index
```

```python
data.sample(1)
```

```python

''.join(str(x) for x in data.iloc[1,0:25])
```

```python
#joining all the sentences in a row and creating a list of it
headlines = []
for row in range(0,len(data.index)):
    headlines.append(''.join(str(x) for x in data.iloc[row,0:25]))
```

```python
headlines[3]
```

```python
#CountVectorizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
countvector = CountVectorizer(ngram_range=(2,2))
traindataset = countvector.fit_transform(headlines)
```

```python
randomforest = RandomForestClassifier(n_estimators = 200, criterion='entropy')
randomforest.fit(traindataset,train['Label'])
```

```python
#Predict for test dataset
test_transform = []
for row in range(0,len(test.index)):
    test_transform.append(''.join(str(x) for x in test.iloc[row,2:27]))
    
```

```python
test_dataset = countvector.transform(test_transform)
predictions = randomforest.predict(test_dataset)
```

```python
#import library to check acccuracy
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
matrix = confusion_matrix(test['Label'],predictions)
print(matrix)
score = accuracy_score(test['Label'],predictions)
print(score)
report = classification_report(test['Label'],predictions)
print(report)


```
